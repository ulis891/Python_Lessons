```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from sympy import *
from sympy.plotting import plot
```

```python
x = list(range(-200, 200))
fx = [-12*i**4*np.sin(np.cos(i))-18*i**3+5*i**2+10 for i in x]
```

```python
plt.plot(x, fx)
plt.show()
```

```python
x = symbols('x', real=True)
```

```python
ex = -12*x**4*sin(cos(x))-18*x**3+5*x*2+10
```

```python
plot(ex, line_color='red')
```

```python
simplify(ex)
```

```python
diff(ex, x)
```

```python
solveset(ex,x)
```
